# Time-Series Analysis of Financial Data Using Stochastic Smoothing Techniques

This project analyzes stock price data for Reliance and TCS using Python.  
The objective is to study trends, fluctuations, and volatility in financial time-series data using smoothing techniques like moving averages.


```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
reliance = yf.download("RELIANCE.NS", start="2022-01-01", end="2024-01-01")
tcs = yf.download("TCS.NS", start="2022-01-01", end="2024-01-01")
```

```python
reliance.columns = ["Close", "High", "Low", "Open", "Volume"]
tcs.columns = ["Close", "High", "Low", "Open", "Volume"]
```

```python
## Reliance Analysis
```

```python
plt.figure(figsize=(12,5))
plt.plot(reliance["Close"], label="Close Price")
plt.title("Reliance Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
reliance["MA_10"] = reliance["Close"].rolling(window=10).mean()

plt.figure(figsize=(12,5))
plt.plot(reliance["Close"], label="Close Price")
plt.plot(reliance["MA_10"], label="10-day MA")
plt.title("Reliance Moving Average")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
## TCS Stock Analysis
```

```python
plt.figure(figsize=(12,5))
plt.plot(tcs["Close"], label="Close Price")
plt.title("TCS Stock Price")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
tcs["MA_10"] = tcs["Close"].rolling(10).mean()

plt.figure(figsize=(12,5))
plt.plot(tcs["Close"], label="Close Price")
plt.plot(tcs["MA_10"], label="10-day MA")
plt.legend()
plt.show()
```

```python
tcs["Volatility"] = tcs["Close"].rolling(10).std()

plt.figure(figsize=(12,5))
plt.plot(tcs["Volatility"], label="Volatility")
plt.legend()
plt.show()
```

```python
## Comparative Analysis
```

```python
plt.figure(figsize=(12,5))
plt.plot(reliance["Close"], label="Reliance")
plt.plot(tcs["Close"], label="TCS")
plt.legend()
plt.title("Price Comparison")
plt.show()
```

```python
rel_norm = reliance["Close"] / reliance["Close"].iloc[0]
tcs_norm = tcs["Close"] / tcs["Close"].iloc[0]

plt.figure(figsize=(12,5))
plt.plot(rel_norm, label="Reliance")
plt.plot(tcs_norm, label="TCS")
plt.legend()
plt.title("Normalized Growth Comparison")
plt.show()
```

```python
reliance["Volatility"] = reliance["Close"].rolling(10).std()
tcs["Volatility"] = tcs["Close"].rolling(10).std()
```

```python
plt.figure(figsize=(12,5))

plt.plot(reliance["Volatility"], label="Reliance")
plt.plot(tcs["Volatility"], label="TCS")

plt.legend()
plt.title("Volatility Comparison")
plt.show()
```

## Key Insights

- Moving averages smooth short-term fluctuations and help reveal the underlying trend in stock prices.
- TCS exhibits relatively stronger growth compared to Reliance when prices are normalized to a common baseline.
- Volatility analysis shows that both stocks experience periods of higher fluctuation, with noticeable spikes indicating market uncertainty.
- The stock price data shows noisy behavior, and smoothing techniques like moving averages help in distinguishing signal from stochastic fluctuations.
